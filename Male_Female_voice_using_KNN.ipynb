```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('voice.csv')
```

```python
data = pd.get_dummies(df)
```

```python
data.head()
```

```python
data_x = data.iloc[:,:20]
data_y = data.iloc[:,20:]
```

```python
data_y.head()
```

```python
data_x.describe()
```

```python
std_scaler = StandardScaler()
df_scale = pd.DataFrame(std_scaler.fit_transform(data_x), columns = data_x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_scale, data_y, test_size = 0.2, random_state=43)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = neigh = KNeighborsClassifier()
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score



k_values = list(range(1, 31, 2)) 
accuracies = []

for k in k_values:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn_model, x_train, y_train, cv=10, scoring='accuracy')
    accuracies.append(scores.mean())


# Plotting accuracy scores versus k values
plt.figure(figsize =(15,8))
plt.plot(k_values, accuracies, marker='o')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.title('Accuracy Scores for Different k values')
plt.grid(True)
plt.show()
```
